```python
import pandas as pd
import random

# Initialize an empty DataFrame
df = pd.DataFrame(columns=["Name", "Gender", "Age", "Height", "Weight"])

for _ in range(5):  # Added more random rows
    name = ''.join(random.choice('abcdefghijklmnopqrstuvwxyz') for _ in range(5))
    gender = random.choice(["M", "F"])
    age = random.randint(18, 60)
    height = random.randint(150, 200)
    weight = random.uniform(50, 100)

    # Append the generated data to the DataFrame
    df = df.concat({"Name": name, "Gender": gender, "Age": age, "Height": height,
                    "Weight": weight})# ignore_index=True)

#Display the dataFrame
print(df)
```
