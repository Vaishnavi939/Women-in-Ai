```python
m1 = np.array([[2,1],[3,4]]) #solution 2x+y=8 3x+4y=18
co_m2 = np.array([8,18])
solution = np.linalg.solve(m1, co_m2)
solution
```

```python
#solve 3x+4y=10, x+y=5
m1 = np.array([[3,4],[1,1]])
co_m2 = np.array([10,5])
solution = np.linalg.solve(m1, co_m2)
solution
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df= pd.read_csv("Amazon Sales Data.csv",encoding='unicode_escape')
```

```python
df
```

```python
#Drop empty columns, status & unnamed
df.drop(index=1)
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
import seaborn as sns
ax = sns.countplot(x='Gender', data=df)
for bars in ax.containers:
  sns.bar_label(bars)

```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum()
print(sales_gen)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
Sales_state = df.groupby('State', as_index=False)['Amount'].sum().sort_values(ascending=False,by="Amount").head()
print(Sales_state.head())  # Just for checking the first few rows

sns.barplot(x='State', y='Amount', data=Sales_state)
plt.show()
```
