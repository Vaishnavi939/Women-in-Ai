```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
Label = LabelEncoder()
x[:, 3] = Label.fit_transform(x[:, 3])
print(x)
```

```python
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('encoder',OneHotEncoder(),[3])],remainder='passthrough')
x=np.array(ct.fit_transform(x))
print(x)
```

```python
a = "encrypt this message using python"
encrypted_message=" "
for i in a:
  if i.isalpha():
    encrypted_message+=chr(219-ord(i))
  elif i==" ":
    encrypted_message+="**"
  else:
    encrypted_message+=i
print(encrypted_message)
```

```python
message="vmxibkg*gsrh*fhrmt*kbgslm!"
decrypted_message=""
for i in message:
  if i.isalpha():
    decrypted_message+=chr(219-ord(i))
  elif i==" ":
    decrypted_message+="**"
  else:
    decrypted_message+=i
print(decrypted_message)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=14)
```

```python
print(x.shape)
print(y.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l=l.fit(x_train,y_train)
l
```

```python
y_pred=l.predict(x_test)
y_pred
```

```python
y_test
```

```python
result=pd.concat([pd.DataFrame(y_pred,columns=['Prediction']),
                  pd.DataFrame(y_test,columns=['Test'])],axis=1)
result['Difference']=result['Prediction']-result['Test']
result
```
