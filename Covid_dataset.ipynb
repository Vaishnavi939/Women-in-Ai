```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_covid=pd.read_excel('/content/Covid_Data.xlsx')
df_covid
```

```python
df_covid.State.value_counts()
```

```python
df_covid.State.value_counts().head(6)
```

```python
df_new=df_covid.groupby('State')['Active Cases'].sum()
df_final=df_new.sort_values(ascending=False).head(5)
df_final
```

```python
df_new=df_covid.groupby('State')['Cured'].sum()
df_final=df_new.sort_values(ascending=False).head(5)
df_final
sns.barplot(data=df_final)
plt.rcParams['figure.figsize']= (10, 20)
plt.title('Active Cases vs State')
plt.xlabel("Active Cases")
plt.ylabel("State")
plt.show()
```

```python
df_new=df_covid.groupby('State')['Active Cases'].sum()
df_new=df_new.sort_values(ascending=False).head(5)
sns.barplot(data=df_new)
```

```python
plt.plot(df_covid["State"],df_covid["Active Cases"])
plt.rcParams['figure.figsize']= (10, 15)
plt.show()
```
